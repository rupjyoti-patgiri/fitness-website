```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as mp
```

```python
df = pd.read_csv('/content/weight_change_dataset.csv')
```

```python
df = df.drop(['Participant ID', 'Stress Level', 'Final Weight (lbs)', 'BMR (Calories)','Weight Change (lbs)','Duration (weeks)'], axis=1)
```

```python
df.isnull()
```

```python
print(df['Physical Activity Level'].unique())
```

```python
activity_levels = df['Physical Activity Level'].unique()
for level in activity_levels:
    df[level] = (df['Physical Activity Level'] == level).astype(int)
df = df.drop('Physical Activity Level', axis=1)
df.head()
```

```python
df
```

```python
activity_levels = df['Sleep Quality'].unique()
for level in activity_levels:
    df[level] = (df['Sleep Quality'] == level).astype(int)
df = df.drop('Sleep Quality', axis=1)
df.head()
```

```python
df
```

```python
df['Gender'] = df['Gender'].map({'M': 0, 'F': 1})
```

```python
df
```

```python
X = df.drop('Daily Caloric Surplus/Deficit', axis=1)
Y = df['Daily Caloric Surplus/Deficit']
```

```python
X
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.metrics import accuracy_score

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)

print("R^2 Score:", r2)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

```python
input_order = [
    "Age",
    "Gender",
    "Current Weight (lbs)",
    "Daily Calories Consumed",
    "Sedentary",
    "Very Active",
    "Lightly Active",
    "Moderately Active",
    "Excellent",
    "Good",
    "Fair",
    "Poor"
]

feature_mapping = {
    "Age": "Age",
    "Gender": "Gender",
    "Current Weight (lbs)": "Current Weight (lbs)",
    "Daily Calories Consumed": "Daily Calories Consumed",
    "Sedentary": "Sedentary",
    "Very Active": "Very Active",
    "Lightly Active": "Lightly Active",
    "Moderately Active": "Moderately Active",
    "Excellent": "Excellent",
    "Good": "Good",
    "Fair": "Fair",
    "Poor": "Poor"
}
```

```python
def collect_user_input():
    user_inputs = {}
    print("Please enter the following details:")

    for feature in input_order:
        while True:
            try:
                if feature in ["Gender", "Sedentary", "Very Active", "Lightly Active",
                              "Moderately Active", "Excellent", "Good", "Fair", "Poor"]:
                    value = int(input(f"Enter {feature} (0 or 1): "))
                    if value not in [0, 1]:
                        raise ValueError
                elif feature == "Age":
                    value = int(input(f"Enter {feature} (e.g., 23): "))
                    if value <= 0:
                        raise ValueError
                elif feature in ["Current Weight (lbs)", "Daily Calories Consumed"]:
                    value = float(input(f"Enter {feature} (numeric value): "))
                    if value < 0:
                        raise ValueError
                else:
                    value = input(f"Enter {feature}: ")

                user_inputs[feature] = value
                break

            except ValueError:
                print(f"Invalid input for {feature}. Please enter a valid value.")

    return user_inputs

user_input = collect_user_input()

mapped_inputs = {}
for user_feature, model_feature in feature_mapping.items():
    mapped_inputs[model_feature] = user_input[user_feature]

input_df = pd.DataFrame([mapped_inputs])
input_df = input_df[feature_mapping.values()]

print("\nPrepared Input for Prediction:")
print(input_df)

predicted_y = model.predict(input_df)

print("\nPredicted Output Y:", predicted_y[0])
```
